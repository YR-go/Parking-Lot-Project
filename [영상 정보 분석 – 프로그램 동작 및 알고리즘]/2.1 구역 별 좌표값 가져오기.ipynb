## 2.1 구역 별 좌표값 가져오기

```python
 def import_value(lot, value):
    x1 = value[lot][0][0]
    y1 = value[lot][0][1]
    x2 = value[lot][1][0]
    y2 = value[lot][1][1]
    lot_value = [x1, y1, x2, y2]
 
   return lot_value
```

각 주차구역 별로 대각선의 양 극 좌표값을 저장한다
