## 2.3 Full / Free 판별

```python
def run(case, value):

    image_origin = cv2.imread(os.path.join(BASE_PATH + str(case)+ str(case) +'.jpg'),cv2.IMREAD_COLOR)   # 주차장 전체사진

    result = []   # 결과값 담길 리스트

    model = load_model('model.h5')
    lot=0
    while lot < 28:
        get_crop_images(image_origin,lot,value,case)
```

- lot 이 0부터 27까지 증가하며 28개 구역별로 image를 자른다.

```python
       test_img = img_to_array(load_img(os.path.join('C:/project/parking/'+str(case) + '_' + str(lot) + '.jpg'), target_size=(224, 224)))
        test_input = preprocess_input(np.expand_dims(test_img.copy(), axis=0))

        global empty
        empty = 0
        pred = model.predict(test_input)
        if pred[0][0] > pred[0][1] : #full>free ->차있다
           result.append(0)
           lot = lot + 1
           print(lot,"번쨰칸)","■")
        elif pred[0][0] < pred[0][1] :#full<free ->비어있다
             result.append(1)
             lot = lot+1
             empty = empty +1
             print(lot,"번쨰칸)","□")

    print('=======================================')
    return image_origin
```

- 위에서 만든 model로 prediction을 진행한다. 
- 그 결과
  - pred[0][0] > pred[0][1] 일 경우#full>free -> 차있다 (■). 
  - pred[0][0] < pred[0][1] 일 경우 #full<free -> 비어있다 (□). 
- 판별 결과를 부등호로(상대비교) 결정하게 한 이유는 다음과 같다. 예를 들어 주차구역이 비어있는데 free의 확률이 1이 아닐 경우. 즉 full 과 free의 확률이 별로 차이가 나지 않 을 경우 정답 class의 확률값이 더 크다는 점을 이용하여 판별 error를 줄이고자 그렇게  설정해두었다.
