## 2.2 주차구역 별로 이미지 자르기

```python
def get_crop_images(image, lot, value, case):
    lot_value = import_value(lot, value)
    x1 = lot_value[0]  # 좌표 순서주의
    y1 = lot_value[1]
    x2 = lot_value[2]
    y2 = lot_value[3]

    dst = image.copy()
    dst = dst[y1:y2, x1:x2]  # [y축, x축]
    
    cv2.imwrite(str(case) + '_' + str(lot) + '.jpg', dst)
```

- import_value()를 호출하여 좌표값을 return 받은 후에 좌표값에 따라 구역 별로     이미지가 잘린다. 총 28번 호출되며 28개의 잘린 이미지가 생성된다. 
- 잘린 이미지는 ‘주차장(num)_lot(num).jpg’ 로 저장된다.
   ex) 첫 번째 주차장 사진에서 23번째 구역 --> ‘1_23.jpg’
- 이후 prediction에서 image를 load하여 array로 바꾸는 과정에 잘린 이미지들이     model에 맞게 target size로 resizing 된 후 load 되는 과정이 필요해서 잘린 이미지를 영상으로 저장하고 다시 load하는 과정을 거친다.
