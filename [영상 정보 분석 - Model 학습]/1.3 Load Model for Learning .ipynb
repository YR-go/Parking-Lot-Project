## 1.3 Load Model for Learning 

```python
base_model =
 MobileNetV2( input_shape=(224,224,3), weights='imagenet', include_top=False )

x = base_model.output
c = GlobalAveragePooling2D()(x)
output = Dense(2, activation='softmax')(c)

model = Model(inputs=base_model.input, outputs=output)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
```

 - Convolution
  - 여러 filter를 사용하여 인접한 pixel간 정보를 분석, 즉 image의 특징 추출
  - MobileNet V2 사용
 - Pooling 
  - Convolution 결과 만들어진 Feature Map을 sampling 하여 크기 축소
  - Global Average Pooling 사용

이 두 가지의 과정을 거치면서 model이 학습된다.

- MobileNetV2

 Google 이 만든 Model로 기존의 convolution의 연산량을 줄여 속도를 향상시켰다.
 MobileNetV2 는 다음과 같이 두 convoution 연산을 수행한다.

- Global Average Pooling (GAP)

 GAP은 각 feature map의 평균값을 뽑아낸다. 따라서 global average pooling layer의 최종 output은 single value로 크기가 감소된 feature map이 된다.

 global average pooling layer는 3D array를 input으로 하여 벡터를 리턴한다.
