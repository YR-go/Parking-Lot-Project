## 1.2 Data Generator

```python
train_datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    brightness_range=[0.8, 1.2],
    shear_range=0.01,
    zoom_range=[0.9, 1.1],
    validation_split=0.1,
    preprocessing_function=preprocess_input
)

val_datagen = ImageDataGenerator(
    validation_split=0.1,
    preprocessing_function=preprocess_input
)
```

- fuction preprocess_input을 사용하여 data preprocessing.
- 90%를 training data 10%를 test data로 설정. 

```python
train_gen = train_datagen.flow_from_directory(
    BASE_PATH,
    target_size=(224, 224),
    classes=['full', 'free'],
    class_mode='categorical',
    batch_size=10,
    shuffle=True,
    subset='training'
)

val_gen = val_datagen.flow_from_directory(
    BASE_PATH,
    target_size=(224, 224),
    classes=['full', 'free'],
    class_mode='categorical',
    batch_size=10,
    shuffle=False,
    subset='validation'
)
```

- 훈련에 사용할 모델인 MobileNetV2의 조건에 맞추어 dataset 이미지들을 target_size를 (224,224)로 설정해준다.
- class를 ['full', 'free']로 설정 -> 즉 판별 결과가 자동차가 주차되어 있는 경우 [1,0], 비어있는 경우 [0,1]로 나옴.
